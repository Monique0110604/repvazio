```python

temperatura = float(input("Digite a temperatura: "))
pressao = int(input("Digite a pressão sistólica: "))
dor = int(input("Digite o nível de dor (0-10): "))


if temperatura > 39.5 or pressao > 180 or dor >= 9:
    classificacao = "Vermelho (Emergência)"
elif temperatura > 38.5 or pressao > 160 or dor >= 7:
    classificacao = "Laranja (Muito urgente)"
elif temperatura > 37.5 or pressao > 140 or dor >= 4:
    classificacao = "Amarelo (Urgente)"
else:
    classificacao = "Verde (Pouco urgente)"


print(f"\nTemperatura: {temperatura}")
print(f"Pressão: {pressao}")
print(f"Dor: {dor}")
print(f"Classificação: {classificacao}")
```
